```python
# Step 1: Count the frequency of each category
category_counts = df1["CATEGORY*"].value_counts()

# Step 2: Plot the pie chart
#labels=category_counts.index: This uses the category names (or labels) as the labels for each slice.
#autopct='%1.1f%%': This formats the percentage values shown on the chart to one decimal place.
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title("Category Distribution")
plt.show()
```

```python
purpose_count=df1["PURPOSE*"].value_counts()
plt.pie(purpose_count,labels=purpose_count,autopct='%1.1f%%')
plt.title("Purpose Distribution")
plt.show()
```

```python
#A histogram is a graph showing frequency distributions.
category_counts = df1["START*"].value_counts()
plt.hist(category_counts)
plt.title("Category Distribution")
plt.show()
```

```python
# Assuming df1 is your DataFrame and CATEGORY* is the column
category_counts = df1["CATEGORY*"].value_counts()
# Create a bar chart
plt.bar(category_counts.index, category_counts.values)
# Add a title
plt.title("Category Distribution")
# Add labels to x and y axes
plt.xlabel("Categories")
plt.ylabel("Frequency")
# Display the bar chart
plt.show()
```

```python
# Assuming df1 is your DataFrame and MILES* is the column
category_counts = df1["MILES*"].value_counts()
# Create a scatter chart
plt.scatter(category_counts.index, category_counts.values)
# Add a title
plt.title("Miles Distribution")
# Add labels to x and y axes
plt.xlabel("Miles Categories")
plt.ylabel("Frequency")
# Display the scatter chart
plt.show()
```

```python
# Assuming df1 is your DataFrame and 'MILES*' is the column you want to plot
plt.boxplot(df1['MILES*'])
# Add a title
plt.title("Boxplot of Miles")
# Add a label to the y-axis
plt.ylabel("Miles")
# Display the boxplot
plt.show()
```
