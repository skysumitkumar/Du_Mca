# This is the PDA which accepting odd pallindrom string

```python
oddPalindrome = md2mc('''PDA
I : a, #  ; a#  -> I
I : b, #  ; b#  -> I
I : a, a  ; aa  -> I
I : b, b  ; bb  -> I
I : a, b  ; ab  -> I
I : b, a  ; ba  -> I

I : c, a  ; ''  -> q1
I : c, b  ; ''  -> q1

I : c, # ; #   -> F

q1 : a, a ; ''  -> q1
q1 : b, b ; ''  -> q1

q1 : '', # ; #   -> F
''')

DOpdaPalindrome = dotObj_pda(oddPalindrome, FuseEdges=True)
DOpdaPalindrome
```

Animation of above PDA

```python
AnimatePDA(oddPalindrome, FuseEdges=True)
display(HTML('<link rel="stylesheet" href="//stackpath.bootstrapcdn.com/font-awesome/4.7.0/css/font-awesome.min.css"/>'))
```

# This is the PDA which accepts even pallindrom strings

```python
evenPalindrome = md2mc('''PDA
I : a, #  ; a#  -> I
I : b, #  ; b#  -> I
I : a, a  ; aa  -> I
I : b, b  ; bb  -> I
I : a, b  ; ab  -> I
I : b, a  ; ba  -> I

I : a, a  ; ''  -> q1
I : b, b  ; ''  -> q1

q1 : a, a ; ''   -> q1
q1 : b, b ; ''   -> q1


q1 : '', # ; #   -> F
''')

DOpdaPalindrome = dotObj_pda(evenPalindrome, FuseEdges=True)
DOpdaPalindrome
```
